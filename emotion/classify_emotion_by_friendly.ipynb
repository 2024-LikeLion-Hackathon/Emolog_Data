```python
import pandas as pd

df = pd.read_csv("감정의_차원.csv")
emotionset = df
```

```python
lowthan3 = emotionset[emotionset['원형성'] <= 3.3]
emotionset = emotionset[emotionset['원형성'] > 3.3]
```

```python
len(emotionset)
lowthan3.to_csv('원형성_lower_than_3.3',index=False)
```

```python
emotionset.reset_index(drop=True, inplace=True)
emotionset.to_csv('new감정',index=False)
len(emotionset)
```

```python
friendly_higherthan4_25 = emotionset[emotionset['친숙성'] >= 4.25]
friendly_higherthan3_63 = emotionset[emotionset['친숙성'] >= 3.63]
friendly_higherthan3 = emotionset[emotionset['친숙성'] >= 2.92]
```

```python
print(len(friendly_higherthan4_25))
print(len(friendly_higherthan3_63))
print(len(friendly_higherthan3))
```

```python
friendly_higherthan4_25.to_csv('친숙성_4.25이상',index=False)
friendly_higherthan3_63.to_csv('친숙성_3.63이상',index=False)
friendly_higherthan3.to_csv('친숙성_3이상',index=False)
```

```python
except_3_63to4_25 = friendly_higherthan3_63[friendly_higherthan3_63['친숙성'] < 4.25]
except_3to3_63 = friendly_higherthan3[friendly_higherthan3['친숙성'] < 3.63]
```
