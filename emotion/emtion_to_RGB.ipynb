```python
import pandas as pd

# df = pd.read_csv("data\\389개의 감정단어")
# emotionset = df.drop(['원형성', '친숙성'], axis=1)
emotionset = pd.read_csv("data\\새로운_단어.csv")
```

```python
emotionset = emotionset.drop(['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6'], axis=1)
emotionset
```

```python
import numericalization as numer

emo_rgb = []
for i in range(len(emotionset)):
    # print(emotionset['단어'][i])
    valence = emotionset['쾌-불쾌'][i]
    arousal = emotionset['활성화'][i]
    color = numer.rgb(valence, arousal)
    emo_rgb.append(color.numericalization())
```

```python
emo_rgb
```

```python
rgb = pd.DataFrame(emo_rgb)
```

```python
emotion_rgb = pd.concat([emotionset, rgb], axis=1)
```
